# (5)  Membuat struktur data dengan format json, secara manual dan konversi menjadi dataset dalam format csv.

```python
import json
import pandas as pd

# Data JSON manual
data_produk = [
    {
        "nama_produk": "Blouse Wanita Floral",
        "kategori": "Pakaian",
        "harga": 175000,
        "stok": 75
    },
    {
        "nama_produk": "Sepatu Heels Nude",
        "kategori": "Alas Kaki",
        "harga": 350000,
        "stok": 40
    },
    {
        "nama_produk": "Serum Vitamin C",
        "kategori": "Skincare",
        "harga": 120000,
        "stok": 90
    }
]

# Simpan ke file JSON
with open("data_produk.json", "w", encoding="utf-8") as f:
    json.dump(data_produk, f, indent=4, ensure_ascii=False)

# Baca kembali file JSON
with open("data_produk.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Ubah ke CSV
df = pd.DataFrame(data)
df.to_csv("dataset_produk.csv", index=False, encoding="utf-8")
```
