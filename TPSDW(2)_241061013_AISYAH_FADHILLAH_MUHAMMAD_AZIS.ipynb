```python
#Mengimport Library
import pandas as pd
import numpy as np

#Menampilkan dataset
df = pd.read_csv('data_tugas 2.csv')
df.head()
```

```python
df.info()
```

Dataset yang digunakan terdiri dari **205 baris** (observasi) dan **9 kolom** (variabel), yaitu normalized-losses, make, bahan-bakar, jumlah-pintu, body-style, wheel-bases, length, width, dan height. Berdasarkan tipe datanya, variabel dalam dataset ini terbagi menjadi dua jenis, yaitu **variabel kategorik (object)** yang meliputi normalized-losses, make, bahan-bakar, jumlah-pintu, dan body-style, serta **variabel numerik (float64)** yang terdiri dari wheel-bases, length, width, dan height. Hasil pemeriksaan awal juga menunjukkan bahwa terdapat beberapa data yang belum lengkap (missing value) pada sejumlah variabel, sehingga diperlukan proses data cleaning sebelum dilakukan analisis lebih lanjut.

##1. Menampilkan Jumlah Data Hilang
Lakukan pemeriksaan terhadap dataset untuk mengetahui jumlah **data hilang (missing values)** pada setiap variabel.

**Instruksi:**
*   Tampilkan jumlah missing value pada setiap kolom.
*   Jelaskan variabel mana yang memiliki data hilang paling banyak.

**Output yang diharapkan:**

Tabel jumlah missing value setiap variabel.


```python
# deteksi pola missing non-standar
missing_patterns = ['?', 'NA', 'N/A', '-', '--', '', 'null', 'None']

# konversi ke NaN
df.replace(missing_patterns, np.nan, inplace=True)

# tabel missing value
missing_table = df.isnull().sum().to_frame(name='Jumlah Missing')
missing_table.sort_values(by='Jumlah Missing', ascending=False)
```

Hasil pemeriksaan missing value menunjukkan bahwa variabel normalized-losses memiliki jumlah data hilang paling banyak, yaitu sebanyak 41 data, diikuti oleh jumlah-pintu, bahan-bakar, dan body-style, dst.

```python
# Mengubah 'normalized-losses' ke numerik agar dapat diproses lebih lanjut,
# karena sebelumnya masih bertipe object akibat adanya nilai non-standar.

df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')

df.info()
```

##2. Mengganti Missing Value pada Variabel Kategori
Untuk variabel berbentuk **kategori**, data hilang harus diganti dengan kategori baru yaitu **"unknown"**.

**Variabel kategorik pada dataset:**
*   make
*   bahan-bakar
*   jumlah-pintu
*   body-style

**Instruksi:**
*   Identifikasi missing value pada variabel kategori
*   Ganti nilai tersebut dengan **"unknown"**

**Output yang diharapkan:**

Tabel ringkasan jumlah setiap kategori pada keempat variabel


```python
# Mengecek jumlah missing value pada variabel kategorik
kategori_cols = ['make', 'bahan-bakar', 'jumlah-pintu', 'body-style']

df[kategori_cols].isnull().sum()
```

```python
# Mengganti missing value pada variabel kategorik dengan "unknown"
df[kategori_cols] = df[kategori_cols].fillna('unknown')
```

```python
# Memastikan bahwa seluruh missing value pada variabel kategorik
# telah berhasil digantikan dengan "unknown"
df[kategori_cols].isnull().sum()
```

```python
# Menampilkan distribusi setiap kategori untuk melihat perubahan data
# setelah proses penggantian missing value
for col in kategori_cols:
  print(f"\nDistribusi {col}:")
  print(df[col].value_counts())
```

##3. Mengganti Missing Value pada Variabel Numerik
Untuk variabel numerik yang memiliki data hilang, lakukan **imputasi menggunakan nilai rata-rata (mean)**.

**Variabel numerik pada dataset:**
*   leght
*   width
*   height

**Instruksi:**
*   Hitung nilai rata-rata masing-masing variabel
*   Gunakan nilai tersebut untuk mengganti data yang hilang.




```python
# Mengecek jumlah missing value pada variabel numerik
numerik_cols = ['length', 'width', 'height']
df[numerik_cols].isnull().sum()
```

```python
# Menghitung nilai rata-rata (mean) dari setiap variabel numerik
mean_length = df['length'].mean()
mean_width = df['width'].mean()
mean_height = df['height'].mean()

# Mengganti missing value pada variabel numerik dengan nilai rata-rata
df['length'] = df['length'].fillna(mean_length)
df['width'] = df['width'].fillna(mean_width)
df['height'] = df['height'].fillna(mean_height)

# Memastikan tidak ada lagi missing value pada variabel numerik
df[numerik_cols].isnull().sum()
```

##4. Min-Max Normalization
Lakukan **Min-Max Normalization** pada variabel: **wheel-bases**


```python
# Mengecek apakah masih terdapat missing value pada variabel wheel-bases
df['wheel-bases'].isnull().sum()
```

```python
# Mengganti missing value pada variabel wheel-bases dengan nilai rata-rata
df['wheel-bases'] = df['wheel-bases'].fillna(df['wheel-bases'].mean())
```

```python
# Memastikan tidak ada lagi missing value pada variabel wheel-bases
df['wheel-bases'].isnull().sum()
```

```python
# Melakukan normalisasi Min-Max pada variabel wheel-bases
# agar nilai berada dalam rentang 0 hingga 1
min_val = df['wheel-bases'].min()
max_val = df['wheel-bases'].max()
df['wheel-bases-normalized'] = (df['wheel-bases'] - min_val) / (max_val - min_val)
```

```python
# Menampilkan perbandingan nilai sebelum dan sesudah normalisasi
df[['wheel-bases', 'wheel-bases-normalized']].head()
```

##5. Z-Score Normalization
Lakukan **Z-score normalization** pada variabel berikut:
*   leght
*   width
*   height


```python
# Melakukan normalisasi Z-score pada variabel numerik
# untuk mengubah distribusi data menjadi memiliki mean = 0 dan standar deviasi = 1
numerik_cols = ['length', 'width', 'height']

for col in numerik_cols:
    mean = df[col].mean()
    std = df[col].std()

    df[col + '_zscore'] = (df[col] - mean) / std

# Menampilkan perbandingan nilai sebelum dan sesudah normalisasi Z-score
df[['length', 'length_zscore',
    'width', 'width_zscore',
    'height', 'height_zscore']].head()
```

##6. Pengolahan Variabel Tambahan
Tahapan ini merupakan preprocessing tambahan di luar instruksi tugas, dilakukan karena variabel **normalized-losses** memiliki jumlah missing value paling tinggi dan merepresentasikan tingkat risiko kendaraan yang penting dalam analisis. Oleh karena itu, dilakukan **imputasi menggunakan mean** agar data tetap lengkap tanpa menghilangkan informasi yang relevan.


```python
# Mengecek apakah masih terdapat missing value pada variabel normalized-losses
df['normalized-losses'].isnull().sum()
```

```python
# Mengganti missing value pada variabel normalized-losses dengan nilai rata-rata
df['normalized-losses'] = df['normalized-losses'].fillna(df['normalized-losses'].mean())
```

```python
# Memastikan tidak ada lagi missing value pada variabel normalized-losses
df['normalized-losses'].isnull().sum()
```
