##1. Import Library

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

Konstruksi model regresi sederhana pada diatas.
Selidiki nilai MAE, RMSE dan MSE
Buat scatter plot regresi, berikan penjelasan dan kaitkan dengan nilai MAE, RMSE dan MSE

##2. Membaca Dataset

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/GeostatsGuy/GeoDataSets/master/1D_Porosity.csv") # data from Dr. Pyrcz's github repository
print(df.head())
```

##3. Informasi Dataset

```python
df.info()
```

##4. Statistik Deskriptif

```python
print(df.describe())
```

##5. Visualisasi Hubungan Variabel

```python
sns.scatterplot(x='Depth', y='Nporosity', data=df)
plt.show()
```

##6. Regresi Linear Sederhana

(1) Menentukan Variabel

```python
X = df[['Depth']]
y = df['Nporosity']
```

(2) Membuat Model Regresi

```python
model = LinearRegression()
model.fit(X, y)
```

(3) Melihat Koefisien Model

```python
print('Intercept:', model.intercept_)
print('Koefisien:', model.coef_)
```

(4) Membuat Visualisasi Model

```python
y_pred = model.predict(X)
print(y_pred)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Depth', y='Nporosity', data=df, label='Data Asli')
plt.plot(X, y_pred, color='red', label='Garis Regresi')
plt.title('Visualisasi Model Regresi Linier Sederhana')
plt.xlabel('Depth')
plt.ylabel('Nporosity')
plt.legend()
plt.grid(True)
plt.show()
```

Dari scatter plot Model Regresi Linear Sederhana di atas terlihat garis regresinya naik, sehingga dapat disimpulkan bahwa ketika depth meningkat maka porosity juga cenderung meningkat. Namun titik-titik data tidak semuanya tepat berada pada garis regresi. Hal ini menunjukkan bahwa hasil model hanya mendekati pola data yang ada. Nilai MAE, MSE, dan RMSE di bawah ini juga menunjukkan adanya perbedaan antara hasil model dan data sebenarnya, walaupun tidak terlalu besar.

##7. Prediksi

Mean Absolute Error (MAE) , MEan Squared Error (MSE) dan Root Mean Square Error (RMSE)


```python
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))


print('MAE:', mae)
print('MSE:',mse)
print('RMSE:',rmse)
```
